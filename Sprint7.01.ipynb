```python
#- Exercici 1
#Calculadora de l'índex de massa corporal

#Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades. Pots obtenir més informació del seu càlcul en:

#Indice de masa corporal IMC que es y como se calcula.

#La funció ha de classificar el resultat en les seves respectives categories

# Función para calcular el IMC

def calcular_imc(peso, altura):
    return peso / (altura ** 2)

# Función para clasificar el IMC por rango de peso correspondiente
def clasificar_imc(imc):
    if imc < 18.5:
        return "Bajo peso"
    elif 18.5 <= imc < 24.9:
        return "Peso normal"
    elif 25 <= imc < 29.9:
        return "Sobrepeso"
    else:
        return "Obesidad"

# Función principal
def main():
    try:
        # Pedir al usuario su altura en metros y su peso en kilogramos
        altura = float(input("Introduce tu altura en metros (ej. 1.75): "))
        peso = float(input("Introduce tu peso en kilogramos (ej. 70): "))
        
        # Calcular el IMC
        imc = calcular_imc(peso, altura)
        
        # Clasificar el IMC
        clasificacion = clasificar_imc(imc)
        
        # Mostrar el resultado
        print(f"Tu IMC es: {imc:.2f}")
        print(f"Clasificación: {clasificacion}")
    except ValueError: #en caso de que el usuario escriba caracteres invalidos, saldra el print de abajo
        print("Por favor, introduce valores numéricos válidos para el peso y la altura.") 

# Ejecutar la función principal
if __name__ == "__main__":
    main()
```

```python
#- Exercici 2
#Convertidor de temperatures.

#Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re). Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions.

def conversion_temperatura(celsius):

    kelvin = celsius + 273.15
    fahrenheit = celsius * 9/5 + 32

    return {"Kelvin": kelvin, "Fahrenheit": fahrenheit} #diccionario de resultados

celsius = float(input("Introduce la temperatura en grados Celsius: ")) #esta es el prompt que vamos a poner arriba hara que el usuario introduzca una temperatura en celsius, la cual sera convertida a K y F.
resultados = conversion_temperatura(celsius) 

print(f"Temperatura en Kelvin: {resultados['Kelvin']}")
print(f"Temperatura en Fahrenheit: {resultados['Fahrenheit']}")
```

```python
#- Exercici 3
#Comptador de paraules d'un text.

#Escriu una funció que donat un text, mostri les vegades que apareix cada paraula.

import re #re es para que no cuente caracteres especiales como palabras
from collections import Counter #facilita la contabilidad de elementos en un iterable.

def contar_frecuencia_palabras(texto):
    # Convertir el texto a minúsculas, ya que sino Hola y hola se contarían como dos palabras distintas
    texto = texto.lower()
    
    # Elimina caracteres especiales
    texto = re.sub(r'[^\w\s]', '', texto)
    
    # Divide el texto en palabras
    palabras = texto.split()
    
    # Utiliza Counter para contar la frecuencia de aparicion de cada palabra
    frecuencia_palabras = Counter(palabras)
    
    return frecuencia_palabras

# Pedimos al usuario que introduzca un texto

texto = input("Introduce un texto: ") 

# Aquí se cuenta la frecuencia de palabras en el texto
frecuencia_palabras = contar_frecuencia_palabras(texto)

# Y aquí, el resultado
print("Frecuencia de palabras en el texto:")
for palabra, frecuencia in frecuencia_palabras.items(): #items devuelve una lista de los pares clave valor (palabras y veces que aparecen)
    print(f"{palabra}: {frecuencia}")
```

```python
#- Exercici 4
# Diccionari invers.

# Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, 
# intercanviats les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment.

def invertir_diccionario(diccionario):  #esta sera la funcion que repase los valores del bucle for uno a uno, invirtiendo los valores de los mismos
    nuevo_diccionario = {}
    duplicados = False

    for clave, valor in diccionario.items():
        if valor in nuevo_diccionario:
            print(f"Error: Múltiples claves ('{nuevo_diccionario[valor]}' y '{clave}') para el mismo valor '{valor}'.")
            duplicados = True
        else:
            nuevo_diccionario[valor] = clave

    if not duplicados:
        return nuevo_diccionario
    else:
        return None


diccionario_1 = {
    'a': 1,
    'b': 2,
    'c': 1,  # Esto causará un error porque el valor '1' ya está en uso
    'd': 4
}


diccionario_2 = {
    'a': 1,
    'b': 2,
    'c': 3,  
    'd': 4
}

#caso 1, ejemplo de error por duplicado

diccionario_invertido = invertir_diccionario(diccionario_1)

if diccionario_invertido:
    print("Diccionario invertido:", diccionario_invertido)
else:
    print("El diccionario no se puede invertir debido a duplicados.")

#caso 2, ejemplo de inversion correcta

diccionario_invertido = invertir_diccionario(diccionario_2)
    
if diccionario_invertido:
    print("Diccionario invertido:", diccionario_invertido)
else:
    print("El diccionario no se puede invertir debido a duplicados.")
```
